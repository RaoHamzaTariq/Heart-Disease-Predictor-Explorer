```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'data_input'

# Define mappings for categorical variables
sex_mapping = {1: 'Male', 0: 'Female'}
cp_mapping = {0: 'Typical angina', 1: 'Atypical angina', 2: 'Non-anginal pain', 3: 'Asymptomatic'}
restecg_mapping = {0: 'Normal', 1: 'ST-T wave abnormality', 2: 'Probable or definite LVH'}
fbs_mapping = {1: 'True', 0: 'False'}
slope_mapping = {0: 'Upsloping', 1: 'Flat', 2: 'Downsloping'}
thal_mapping = {0: 'Normal', 1: 'Fixed defect', 2: 'Reversible defect',3: "Other"}
exang_mapping = {1 : 'Yes', 0:'No'}

# Replace numeric values with text labels
data_input['sex'] = data_input['sex'].replace(sex_mapping)
data_input['cp'] = data_input['cp'].replace(cp_mapping)
data_input['restecg'] = data_input['restecg'].replace(restecg_mapping)
data_input['fbs'] = data_input['fbs'].replace(fbs_mapping)
data_input['slope'] = data_input['slope'].replace(slope_mapping)
data_input['thal'] = data_input['thal'].replace(thal_mapping)
data_input['exang'] = data_input['exang'].replace(exang_mapping)

# Print the updated DataFrame
data_input
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Assuming you have a DataFrame named 'data_input'

# List of categorical columns to encode
categorical_columns = ['sex', 'cp', 'restecg', 'fbs', 'slope', 'thal','exang']

# Initialize LabelEncoder
sex_encoder = LabelEncoder()
cp_encoder = LabelEncoder()
restecg_encoder = LabelEncoder()
fbs_encoder = LabelEncoder()
slope_encoder = LabelEncoder()
thal_encoder = LabelEncoder()
exang_encoder = LabelEncoder()

# Apply LabelEncoder to each categorical column
data_input['sex'] = sex_encoder.fit_transform(data_input['sex'])
data_input['cp'] = cp_encoder.fit_transform(data_input['cp'])
data_input['restecg'] = restecg_encoder.fit_transform(data_input['restecg'])
data_input['fbs'] = fbs_encoder.fit_transform(data_input['fbs'])
data_input['slope'] = slope_encoder.fit_transform(data_input['slope'])
data_input['thal'] = thal_encoder.fit_transform(data_input['thal'])
data_input['exang'] = restecg_encoder.fit_transform(data_input['exang'])
# Print the updated DataFrame
data_input.head()
```

```python
X = data_input.drop('target', axis=1)
Y = data_input['target']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=42)

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
score = model.score(X_test,Y_test)
score
```
